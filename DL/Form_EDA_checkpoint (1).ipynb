# Cleaning Data

```python
#Check for missing values
print("Missing values before cleaning:\n", df.isnull().sum())
```

```python
#Drop rows where 'الاسم' (Name) is missing
df_cleaned = df.dropna(subset=['الاسم'])
```

```python
#Drop rows where'لو فكرت قبل كده ففكره تساعد في انتشار المساحات الخضراء ممكن تشاركها معانا؟' is missing
df_cleaned = df.dropna(subset=['لو فكرت قبل كده ففكره تساعد في انتشار المساحات الخضراء ممكن تشاركها معانا؟'])
```

```python
#Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()
```

```python
#Re-check for missing values after cleaning
print("Missing values after cleaning:\n", df_cleaned.isnull().sum())
```
